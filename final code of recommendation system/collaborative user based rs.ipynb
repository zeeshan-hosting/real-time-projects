```python
df1 = pd.read_csv("/content/drive/MyDrive/kaggle/unique_tracks.txt",sep='<SEP>',
                   names=["track_id","song_id","artist_name","release"])
df1.head()
```

```python
df2 = pd.merge(df, df1, how='left', on='song_id')
df2.drop(['track_id'], axis=1, inplace=True)
df2 = df2.sample(n=100000, random_state=1)
df2.head()
```

```python
userRatings = df2.pivot_table(index=['user_id'],columns=['artist_name'],values='ratings')
userRatings.head()
# print("Before: ",userRatings.shape)
```

```python
userRatings = userRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
#userRatings.fillna(0, inplace=True)
userRatings.head()
# print("After: ",userRatings.shape)
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix.head(10)
```

```python
corrMatrix.shape
```

```python
def get_similar(song_name, rating):
    similar_ratings = corrMatrix[song_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
inp = [("4 Non Blondes",4),("Blur",3),("Bloc Party",1),("311",2),("Whitney Houston",5)]
similar_songs = pd.DataFrame()
for song,rating in inp:
    similar_songs = similar_songs.append(get_similar(song,rating),ignore_index = True)

similar_songs.head(10)
```
